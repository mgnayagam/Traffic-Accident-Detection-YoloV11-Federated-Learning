# Testing the model

```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/Traffic_Accident_Project/runs/classify/train/weights/best.pt")
```

```python
import os

test_folder = "/content/dataset/test/Accident"
image_name = os.listdir(test_folder)[0]

img_path = os.path.join(test_folder, image_name)

result = model.predict(img_path)

pred_class = result[0].names[result[0].probs.top1]
confidence = result[0].probs.top1conf.item()

print("Image:", image_name)
print("Prediction:", pred_class)
print("Confidence:", f"{confidence:.2%}")
```

```python
import os

test_folder = "/content/dataset/test/Non Accident"
image_name = os.listdir(test_folder)[0]

img_path = os.path.join(test_folder, image_name)

result = model.predict(img_path)

pred_class = result[0].names[result[0].probs.top1]
confidence = result[0].probs.top1conf.item()

print("Image:", image_name)
print("Prediction:", pred_class)
print("Confidence:", f"{confidence:.2%}")
```

```python
import cv2

video_path = "/content/drive/MyDrive/testing/videoplayback.mp4"
cap = cv2.VideoCapture(video_path)

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_path = f"/content/frame_{frame_count}.jpg"
    cv2.imwrite(frame_path, frame)

    result = model.predict(frame_path, verbose=False)

    pred_class = result[0].names[result[0].probs.top1]
    confidence = result[0].probs.top1conf.item()

    print(f"Frame {frame_count}: {pred_class} ({confidence:.2%})")

    frame_count += 1

cap.release()
```
